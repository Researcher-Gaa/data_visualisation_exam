###  Read Csv Via Panda and Visualize it

```python
sns.set(rc={"figure.figsize":(25,10)})
```

### All Available Causes against Occurance Interval by Population

```python
fullinfo = sns.lineplot(x = "Cause", y = "Upper_CI", data = dataA, hue = "Population", style="Population",ci=False, markers = True)
```

### Searching for the Severity of each Injuries

```python
bardata = sns.barplot(x = "Cause", y = "Upper_CI", data = dataA, hue = "Severity", ci=False)
```

### Comparing Upper and Lower Confident Interval growth as per Causes

```python
scat = sns.scatterplot(x="Lower_CI", y="Upper_CI", data = dataA, hue = "Cause", size ="Cause", sizes=(15,150))
```
